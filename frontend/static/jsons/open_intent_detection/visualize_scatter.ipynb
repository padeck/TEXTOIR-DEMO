```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json('ADB_analysis.json')
```

```python
s_adb_dict = df['banking_ADB_0']['points']
```

```python
len(s_adb_dict)
```

```python
import plotly.express as px
import pandas as pd

# Prepare your data as a DataFrame
data = []
for class_name, coordinates in s_adb_dict.items():
    if coordinates:
        for coord in coordinates:
            data.append([coord[0], coord[1], class_name])
    else:
        print(f'Class "{class_name}" has no coordinates.')

custom_colors = [
    "rgb(31, 119, 180)", "rgb(255, 127, 14)", "rgb(44, 160, 44)",
    "rgb(214, 39, 40)", "rgb(148, 103, 189)", "rgb(140, 86, 75)",
    "rgb(227, 119, 194)", "rgb(127, 127, 127)", "rgb(188, 189, 34)",
    "rgb(23, 190, 207)", "rgb(174, 199, 232)", "rgb(255, 187, 120)",
    "rgb(152, 223, 138)", "rgb(255, 152, 150)", "rgb(197, 176, 213)",
    "rgb(196, 156, 148)", "rgb(247, 182, 210)", "rgb(199, 199, 199)",
    "rgb(219, 219, 141)", "rgb(158, 218, 229)"
]

# Convert to a DataFrame for easier plotting
df = pd.DataFrame(data, columns=["x", "y", "class"])

# Create a scatter plot using Plotly
fig = px.scatter(df, x="x", y="y", color="class", title="Open Intent Detection", 
                 labels={"x": "X Coordinate", "y": "Y Coordinate", "class": "Class"},
                 color_discrete_sequence=custom_colors)

# Show the interactive plot
fig.update_layout(showlegend=True)
fig.show()
```
