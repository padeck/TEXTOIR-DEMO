```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json('ADB_analysis.json')
```

```python
s_adb_dict = df['stackoverflow_ADB_1']['points']
```

```python
# Define a colormap
colors = plt.cm.tab10(range(len(s_adb_dict)))  # Generate colors for each class


# Plot each class
for (class_name, coordinates), color in zip(s_adb_dict.items(), colors):
    if not coordinates:  # Check if the list is empty
        print(f"Warning: No coordinates found for class '{class_name}', skipping.")
        continue
    
    # Extract x and y coordinates from lists
    x, y = zip(*coordinates)  # Unpack the coordinate pairs (x, y)
    plt.scatter(x, y, label=class_name, color=color)

# Add legend and labels
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Scatter Plot of Classes in 2D Space")
plt.legend()
plt.grid(True)
plt.show()
```
